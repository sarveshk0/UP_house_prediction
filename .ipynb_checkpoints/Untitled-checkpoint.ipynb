```python
import pandas as pd

# 1. Pehle comma hatao, 'L' hatao, rupee symbol hatao, etc.
# Total Price cleaning
data['Total Price'] = data['Total Price'].str.replace('L', '')  # L hata diya
data['Total Price'] = data['Total Price'].str.replace(',', '')  # Comma hata diya
data['Total Price'] = pd.to_numeric(data['Total Price'], errors='coerce')  # Convert to float

# 2. Rupee_per_area cleaning
data['Rupee_per_area'] = data['Rupee_per_area'].str.replace(',', '')  # Comma hata diya
data['Rupee_per_area'] = data['Rupee_per_area'].str.replace('/ sq ft', '')  # ' / sq ft' hata diya
data['Rupee_per_area'] = pd.to_numeric(data['Rupee_per_area'], errors='coerce')  # Convert to float

# 3. Ab describe() karo
print(data.describe())
```

```python


# Function to extract BHK number
def extract_bhk(x):
    if pd.isnull(x):
        return None
    if 'Studio' in x:
        return 1   # Studio apartment ko 1 BHK maan rahe hain
    digits = ''.join([ch for ch in x if ch.isdigit()])
    return int(digits) if digits else None

# New column 'BHK'
data['BHK'] = data['Type_of_plot'].apply(extract_bhk)

# Dekhne ke liye
print(data[['Type_of_plot', 'BHK']].head(10))
```

```python
data.head()
```

```python
# 1. Remove NaN from 'Type_of_plot'
data = data.dropna(subset=['Type_of_plot'])


```

```python
data.head()

```

```python
data['Apartement'] = data['Apartement'].str.replace(' BHK Apartment', '', regex=False)
# Remove the 'Type_of_plot' column
data.drop(columns=['Type_of_plot'], inplace=True)

```

```python
data.head()
```

```python
# Step 1: Clean 'Total Price' column properly
data['Total Price'] = data['Total Price'].str.replace(r'[cC][rR]', '', regex=True)  # remove 'cr', 'Cr', 'CR'
data['Total Price'] = data['Total Price'].str.replace(r'[lL]', '', regex=True)      # remove 'l', 'L'
data['Total Price'] = data['Total Price'].str.strip()                               # remove extra spaces

# Step 2: Convert to float
data['Total Price'] = data['Total Price'].astype(float)
```

```python
print(data['Total Price'].unique())
```

```python

# Convert the full column to numeric
data['Apartement'] = pd.to_numeric(data['Apartement'], errors='coerce')
```

```python
# Step 1: Rename the column
data = data.rename(columns={'Rupee_per_area': 'Rupee_per_area(sq-feet)'})

# Step 2: Clean and convert the values
data['Rupee_per_area(sq-feet)'] = (
    data['Rupee_per_area(sq-feet)']
    .str.replace(r'[^0-9.]', '', regex=True)  # remove anything that's NOT a number or decimal
    .astype(float)  # convert to float
)


data.head()
```

```python
data.drop(columns=['Open_Date'], inplace=True)
data.drop(columns=['Description'], inplace=True)
data.drop(columns=['Project Name'], inplace=True)
data.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
data.head()
```

```python
data.drop(columns=['Unit'],inplace=True)
data.head()
```

```python
# Remove all rows that contain any NaN values
data = data.dropna()
```

```python
data.to_csv("Cleaned_data.csv")
```

```python
X=data.drop(columns=['Total Price'])
y=data['Total Price']            
```

```python
pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
# Applying Linaer Regression
```

```python
column_trans = make_column_transformer(
    (OneHotEncoder(sparse_output=False), ['City']),
    remainder='passthrough',
     force_int_remainder_cols=False 
)
```

```python
scaler=StandardScaler()
```

```python
lr=LinearRegression()
```

```python
pipe=make_pipeline(column_trans,scaler,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_lr=pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred_lr)
```

```python
# Applying Lasso
```

```python
lasso=Lasso()
```

```python
pipe=make_pipeline(column_trans,scaler,lasso)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_lasso=pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred_lasso)
```

```python
# Applying Ridge
```

```python
ridge=Ridge()
```

```python
pipe=make_pipeline(column_trans,scaler,ridge)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_ridge=pipe.predict(X_test)
r2_score(y_test,y_pred_ridge)
```

```python
print("No regularization:",r2_score(y_test,y_pred_lr))
print("Lasso:",r2_score(y_test,y_pred_lasso))
print("Ridge",r2_score(y_test,y_pred_ridge))
```
